```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
```

```python
paragraph = ["In the stillness of the night, oh night so deep,",
             "Dreams take flight, in the night, we sleep.",
             "Underneath the silver moon, the moon so bright,",
             "Whispers croon, in the moonlight, all's right."]
```

```python
# Bag Of Word Model
vectorizer = CountVectorizer()
bag_of_words_data = vectorizer.fit_transform(paragraph)
```

```python
bow_df = pd.DataFrame(data=bag_of_words_data.toarray(),columns=vectorizer.get_feature_names_out())
bow_df
```

```python
# TF-IDF Model
vectorizer = TfidfVectorizer()
tf_data = vectorizer.fit_transform(paragraph)
```

```python
tf_df = pd.DataFrame(data=tf_data.toarray(),columns=vectorizer.get_feature_names_out())
tf_df
```
