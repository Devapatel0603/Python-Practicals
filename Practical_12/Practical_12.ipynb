```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# Plot the different districts of the Gujarat State using the geographic data and display the distribution of population along with district boundary using basemap.
data = pd.read_csv('Gujarat_Districts.csv')
data.info()
```

```python
districts = data['District']
populations = data['Population']
coordinates = data['Co-Ordinate']
latitudes, longitudes = zip(*[eval(coord) for coord in coordinates])
```

```python
plt.figure(figsize=(15,10))
m = Basemap(projection='mill',
            llcrnrlat=20,
            urcrnrlat=25,
            llcrnrlon=68,
            urcrnrlon=75,
            resolution='f')
m.drawcoastlines()
m.drawlsmask(land_color='lightgreen',ocean_color='aqua')
m.drawcountries()

x, y = m(longitudes, latitudes)
m.plot(x, y, marker='o', markersize=6,markerfacecolor='red', linewidth=0)

for i in range(len(districts)):
    plt.text(x[i], y[i], f"{districts[i]}\n{populations[i]}", fontsize=8, ha='right')
plt.show()
```
