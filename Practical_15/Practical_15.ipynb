```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv('loan_data_set.csv')
df.head()
```

```python
# a) Perform univariate and multivariate analysis of the attributes of loan dataset. Use binning/box plot/scatter plot/histogram for it.
# Univariate
loan_y_df=df.loc[df['Loan_Status']=='Y']
loan_n_df=df.loc[df['Loan_Status']=='N']

plt.figure(figsize=(7,5))
plt.plot(loan_y_df['ApplicantIncome'],np.zeros_like(loan_y_df['ApplicantIncome']),'o',color="red",label='Y',alpha=0.75)
plt.plot(loan_n_df['ApplicantIncome'],np.zeros_like(loan_n_df['ApplicantIncome']),'o',color="blue",label='N',alpha=0.6)
plt.xlabel('Applicant Income')
plt.legend(loc=0)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(data=loan_y_df, x='ApplicantIncome', kde=True, hue='Loan_Status', palette="Set1",label="Y")
sns.histplot(data=loan_n_df, x='ApplicantIncome', kde=True, hue='Loan_Status', palette="Set2",label="N")
plt.title('Histogram of Applicant Income')
plt.xlabel('Applicant Income')
plt.ylabel('Frequency')
plt.legend(loc=0)
plt.show()
```

```python
# Multivariate
plt.scatter(loan_y_df['ApplicantIncome'],loan_y_df['CoapplicantIncome'],label='Y')
plt.scatter(loan_n_df['ApplicantIncome'],loan_n_df['CoapplicantIncome'],label='N')
plt.xlabel('Applicant Income')
plt.ylabel('Co-applicant Income')
plt.legend(loc=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=df, palette={'Y': 'red', 'N': 'lightblue'})
plt.title('Applicant Income vs. Loan Status')
```

```python
sns.pairplot(df, hue='Loan_Status', palette={'Y': 'red', 'N': 'lightblue'})
plt.show()
```

```python
# b) Find the Correlation among the attributes of the datasets. Apply covariance and correlation to the attributes and display it.
numeric_cols=['ApplicantIncome','CoapplicantIncome','LoanAmount']
sns.heatmap(df[numeric_cols].corr(),annot=True)
plt.title('Heatmap of correlation')
```

```python
sns.heatmap(df[numeric_cols].cov(),annot=True)
plt.title('Heatmap of covariance')
```
