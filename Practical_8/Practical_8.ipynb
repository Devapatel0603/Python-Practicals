```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nodes = ["A","B","C","D","E","F","G"]
edges = [("A","C"),("A","D"),("E","C"),("B","D"),("B","E"),("F","G"),("E","F"),("G","D")]
```

```python
graph = nx.Graph()
```

```python
for n in nodes:
  graph.add_node(n)
for e in edges:
  graph.add_edge(e[0],e[1])
nx.draw(graph,with_labels=True,node_color="skyblue",node_size=1000,font_color="white",width=2)
plt.margins(0.2)
plt.show()
```
